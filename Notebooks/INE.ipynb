### Income

#### Idea

We combined all the files in the Income_data folder with a for loop to build the DataFrame Income which we then cleaned and explored.

```python
import os
```

```python
## CPRO 1 
exc = pd.read_csv('../INE/Income_data/30851.csv', sep=';')
exc['Unidades territoriales'] = exc['Unidades territoriales'].astype(str)
exc = exc[exc['Periodo']==2017]
exc = exc[exc['Indicadores de renta media'] == 'Renta media por persona ']
exc['Total'] = exc['Total'].fillna(method='ffill')

## All the other provinces 
Path = "../INE/Income_data/"
filelist = os.listdir(Path)
location = []
Income = pd.DataFrame()
for i in filelist:
    if i.endswith(".csv") and i != '30851.csv':  # You could also add "and i.startswith('f')
        with open(Path + i, 'r') as f:
            df = pd.read_csv(Path + i, sep=';')
            Income = pd.concat([Income, df], ignore_index=True)
```

```python
## Final Income dataset
Income = pd.concat([Income, exc], ignore_index=True)

Income['Unidades territoriales'] = Income['Unidades territoriales'].astype(str)
Income = Income[Income['Unidades territoriales'].str.contains('|'.join(['sección']))]
Income= Income[Income['Periodo']==2017]
Income = Income[Income['Indicadores de renta media'].str.contains('|'.join(['persona']))].drop(['Periodo','Indicadores de renta media'],axis=1)

Income['Unidades territoriales'] = Income['Unidades territoriales'].str.split(' ')
Income = Income[Income['Total'].notna()]

number=[]
for row in Income['Unidades territoriales']:
    number.append(row[0])
Income['Unidades territoriales']=number

data=[]
Income = Income.reset_index(drop=True)
for index , row in Income.iterrows():
    a= Income['Unidades territoriales'][index][:2]
    b = Income['Unidades territoriales'][index][2:5]
    c = Income['Unidades territoriales'][index][5:7]
    d= Income['Unidades territoriales'][index][7:10]
    data.append((a,b,c,d))
    if Income['Total'][index] == '.':
        Income['Total'][index]= 3.000
        
        
cols=['CPRO','CMUN','CDIS','CSEC']
result = pd.DataFrame(data, columns=cols)
result = result.astype(int)
Income['Total'] = Income['Total'].astype(str).astype(float, 3)*1000
Income = pd.concat([Income,result],axis =1).drop(['Unidades territoriales'],axis =1)

Income.head()
```

#### Idea

Use GeoSeries.centroid to calculate centroid of INE data, match those withshapes.geojson data by summing up all data of the cetroids which are within the multipolygons of shapes['geometry']. 



```python
INE_final_gpd['centroid'] =  INE_final_gpd['geometry'].centroid
INE_final_gpd = INE_final_gpd.drop(['geometry','Shape_area'],axis=1)

INE_final_gpd = gpd.GeoDataFrame(INE_final_gpd, crs="EPSG:4326", geometry= 'centroid')
```
